```python
# 如需增加条目，只需向这个列表怎加属性即可
attr = [["姓名"],
        ["学号"],
        ["性别"],
        ["专业"],
        ["入学年月"],
        ["已获得","毕业时最低应获学分/已获得学分"],
        ["学院"],
        ["毕业年月"]]

```

```python
# 查找属性
def find_str(str, attr):
    '''
    判断attr中的字符串是否在str中出现过，
    如果出现过，返回这个attr中的字符串
    '''
    
    for k in attr:
        for c in k:
            if c in str:
                return (True, k[0])
    return (False, "")
   
def find_attr(attr):
    info = {}
    for row in sheet.iter_rows():
        for cell in row:
            if cell.value == None:
                continue
            strs = cell.value.split('：')
            strs = [''.join(s.split()) for s in strs]
            res, k = find_str(strs, attr)
            if res == True:
                s = cell.value.split('：')[-1]
                info[k] = s
    return info
studen_info = find_attr(attr)
studen_info
```

```python
# 查找成绩
score_titles = ["学习时间","课程名称","备注","学分",
              "成绩","课程性质"]
search_score_title = "课程名称"
def find_cell(str):
    '''
    查找哪几个cell的value完全等于str
    返回坐标集
    '''
    res = []
    for row in sheet.iter_rows():
        for cell in row:
            if cell.value == str:
                res.append((cell.row, cell.column))
    return res

points = find_cell(search_score_title)
points = [(x + 1, y - 1) for (x, y) in points]



def get_lessons_dict(sites):
    '''
        获取所有课程信息，传入所有的起始点信息
        返回课程字典
    '''
    from itertools import zip_longest
    lessons = []
    for (x, y) in sites:
        nx = x
        for row in sheet.iter_rows(min_row=nx):
            if row[y].value == "" or row[y].value == None:
                break
            ny = y
            lesson_info = []
            for i in range(0, len(score_titles) + 1):
                if sheet.cell(nx, ny).value != None:
                    lesson_info.append(sheet.cell(nx, ny).value)
                ny += 1
            lessons.append(dict(zip_longest(score_titles, lesson_info,fillvalue="")))
            nx += 1
    return lessons

lessons_dict = get_lessons_dict(points)
for d in lessons_dict:
    print(d)
```

```python
rank2score = {
    "优":"90",
    "良":"80",
    "及格":"60",
    "免修":"80"
}

# 计算平均分
def calculate_average(lessons_dict, condition):
    '''
    计算平均分，传入成绩字典,和某个课程应该满足的条件
    返回平均值
    '''
    sum = 0
    ratio = 0
    for lesson in lessons_dict:
        if condition(lesson) == True:
            n = 0
            try:
                n = float(lesson["成绩"])
            except ValueError:
                n = float(rank2score[lesson["成绩"]])
            sum += n * float(lesson["学分"])
            ratio += float(lesson["学分"])
    return sum / ratio

def calculate_condition(lesson):
    '''
    接收一个课程字典，判断是否计算这个成绩
    '''
    if lesson["课程性质"] != "":
        return True
    return False

ave = calculate_average(lessons_dict, calculate_condition)
studen_info["平均分"] =  str(ave)
studen_info
```

```python
# 定义要写入的数据
header = ["学号","姓名","性别","学院","专业","已获得","入学年月","毕业年月","平均分"]
header
```

```python
studen_info = {k:studen_info[k] for k in header}
studen_info
```

```python

# 检查文件是否存在
if not os.path.exists('new_file.xlsx'):
    # 如果文件不存在，创建一个新的文件并写入表头和第一行的数据
    wb = Workbook()
    ws = wb.active
    ws.append(header)  # 添加表头
    ws.append(list(studen_info.values()))  # 添加第一行的数据
else:
    # 如果文件存在，加载该文件并追加新的数据
    wb = load_workbook('new_file.xlsx')
    ws = wb.active
    ws.append(list(studen_info.values()))  # 添加新的数据

# 保存修改
wb.save("new_file.xlsx")
```
